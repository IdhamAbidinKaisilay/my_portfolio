#Data Duplicated
Data Duplicated adalah baris data yang muncul lebih dari sekali dalam dataset.

```python
import pandas as pd
```

```python
# Kita punya data awal, berbentuk dictionary
data = {
    'A':[1, 2, 2, 1, 5],
    'B':[5, 6, 6, 8, 9],
    'C':[1, 2, 2, 1, 6]
}
df = pd.DataFrame(data)
df
```

##Mengetahui Data Duplicate

```python
# Cek baris mana yang duplikat dengan fungsi .duplicated()
# -> Python akan membandingkan setiap baris dengan baris sebelumnya
# -> Jika ada yang sama persis, ditandai True
filter = df.duplicated()
df[filter]
# Hasilnya hanya baris index 2 yang muncul,
# karena isinya (2,6,2) sama persis dengan baris index 1.
```

##Mengatasi Data Duplicate

```python
# Menghapus baris duplikat dari seluruh DataFrame
df.drop_duplicates()

# -> Baris index 2 otomatis hilang, karena sudah ada baris yang sama (index 1).
# -> Jadi sisa data hanya baris unik.
```

```python
# Menghapus duplikat tapi hanya berdasarkan kolom tertentu (subset)
# Contoh: subset=['A', 'C']
df.drop_duplicates(subset=['A', 'C'])

# -> Python hanya membandingkan isi kolom A dan C.
# -> Jadi baris index 3 (A=1, C=1) dianggap duplikat dari index 0 (A=1, C=1),
#    walaupun kolom B berbeda (5 vs 8).
# -> Hasil akhirnya: hanya menyisakan baris dengan kombinasi (A,C) yang unik.
```
