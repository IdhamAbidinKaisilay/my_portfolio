```python
import pandas as pd
```

```python
path = '/content/drive/MyDrive/Data Analyst 2025 + SQL + Excel/Portfolio/data/data_dummy_unclean.csv'
df = pd.read_csv(path)
df
```

#Cleaning Data

```python
# Melihat informasi umum dataset: jumlah baris, kolom, tipe data
df.info()
# Insight: Ada 103 baris, 9 kolom, tipe campuran
```

```python
# Ubah kolom "Date of Joining" jadi datetime agar bisa dihitung
# Parameter errors='coerce' → nilai salah otomatis jadi NaT
df['Date of Joining'] = pd.to_datetime(df['Date of Joining'], errors='coerce')
```

```python
df.info()
# Insight: sekarang "Date of Joining" bertipe datetime, bukan object
```

```python
# Cek jumlah nilai kosong (missing value) di setiap kolom
df.isnull().sum()
# Insight: Income ada 16 kosong, City ada 14 kosong
```

```python
# Tampilkan baris yang kolom "Income" masih kosong
df[df['Income'].isnull()]
# Insight: terlihat siapa saja yang datanya belum ada income
```

```python
df['Income'].fillna(0, inplace=True)
# Ganti nilai kosong di Income dengan 0
```

```python
df.isnull().sum()
# Insight: semua nilai kosong Income sekarang jadi 0
```

```python
# Tampilkan baris yang kolom "City" masih kosong
df[df['City'].isnull()]
# Insight: ada 14 baris tanpa nama kota
```

```python
# Ganti nilai kosong di City dengan "Unknown"
df['City'].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum()
# Insight: semua nilai kosong City jadi "Unknown"
```

```python
# Melihat statistik ringkas data numerik (min, max, mean, dll)
df.describe().round(2)
# Insight: Income ada nilai negatif, umur min 25, max 70
```

```python
# Filter data Income yang kurang dari 0
filter_income = df['Income'] < 0
df[filter_income]
# Insight: ada 2 baris dengan income -10000
```

```python
# Ambil index income negatif, lalu ubah nilainya jadi 0
index_income = df[filter_income].index
df.loc[index_income, 'Income'] = 0
```

```python
df[filter_income]
# Insight: income negatif sudah diperbaiki jadi 0
```

```python
# Pisahkan kolom Email_POS jadi dua kolom baru: Email dan Pos
df[['Email', 'Pos']] = df['Email_POS'].str.split(';', expand=True)
df.drop('Email_POS', axis=1, inplace=True)
```

```python
df
# Insight: kolom Email_POS sudah dihapus, diganti Email & Pos
```

```python
# Cek baris duplikat (baris yang sama persis dengan sebelumnya)
df[df.duplicated()]
# Insight: ada 3 baris dianggap duplikat
```

```python
# Hapus baris duplikat agar data tidak ganda
df = df.drop_duplicates()
```

```python
df
# Insight: total baris berkurang dari 103 jadi 100
```
