```python
import pandas as pd
```

```python
path = '/content/drive/MyDrive/Data Analyst 2025 + SQL + Excel/Portfolio/data/data_dummy_unclean.csv'
df = pd.read_csv(path)
df
```

#Cleaning Data

```python
df.info()
```

```python
# rubah tipe data 'Date of Joining' ke datetime (sebelumnya bertipe object)
df['Date of Joining'] = pd.to_datetime(df['Date of Joining'], errors='coerce')
```

```python
df.info() # tipe data 'Date of Joining'
```

```python
# cek nilai null
df.isnull().sum()
```

```python
# ada 2 kolom yang memiliki null, 'Income' & 'City'
# cek nilai null di 'Income'
df[df['Income'].isnull()]
```

```python
# perbaiki nilai null di 'Income'
df['Income'].fillna(0, inplace=True)
```

```python
df.isnull().sum() #cek nilai null di 'Income'
```

```python
# cek nilai null di 'City'
df[df['City'].isnull()]
```

```python
# perbaiki nilai null di 'City'
df['City'].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum() #cek nilai null di 'City'
```

```python
df.describe().round(2)
```

```python
filter_income = df['Income'] < 0
df[filter_income]
```

```python
index_income = df[filter_income].index
df.loc[index_income, 'Income'] = 0
```

```python
df[filter_income]
```

```python
df[['Email', 'Pos']] = df['Email_POS'].str.split(';', expand=True)
df.drop('Email_POS', axis=1, inplace=True)
```

```python
df[df.duplicated()]
```

```python
df = df.drop_duplicates()
```
