#Data Error / Tidak Sesuai
Data tidak sesuai tergolong nilai data yang salah karena pengetikan, nilai data yang tidak wajar dan kolom-kolom yang menjadi satu.

```python
import pandas as pd
```

```python
path = '/content/drive/MyDrive/Data Analyst 2025 + SQL + Excel/Section 10 Pandas: Pembersihan Data/MISSION+7+FILES/biodata_error.csv'
df = pd.read_csv(path)
df

# Dari tabel terlihat beberapa error:
# - Ada umur negatif (-5), terlalu tua (130, 180) → tidak masuk akal
# - Ada jenis kelamin "Alien" (tidak valid)
# - Ada penulisan tidak konsisten ("laki-laki" vs "Laki-laki", "perempuan" vs "Perempuan")
# - Ada kolom "Email dan POS" yang isinya dua data digabung jadi satu → harus dipisah nantinya
```

```python
# Default: hanya tampilkan ringkasan utk kolom numerik (Umur saja)
df.describe()

## Kalau mau semua kolom (numerik & kategori), tambahkan include='all'
## df.describe(include='all')

# Insight:
# - Umur minimum = -5 (tidak mungkin)
# - Umur maksimum = 180 (tidak wajar untuk manusia)
# - Jadi kita perlu perbaikan di kolom ini.
```

```python
# Buat filter untuk cari umur yang tidak masuk akal (<0 atau >100)
filter_umur = (df['Umur'] < 0) | (df['Umur'] > 100)
df[filter_umur]

# Hasil filter menunjukkan Messi, Maradona, dan Ronaldinho punya umur salah.
```

```python
# Ambil index data yang salah, lalu kita ganti nilainya jadi 40 (asumsi rata-rata usia wajar)
index_umur = df[filter_umur].index
df.loc[index_umur, 'Umur'] = 40
df

# Insight:
# - Umur -5 → jadi 40
# - Umur 130 → jadi 40
# - Umur 180 → jadi 40
# - Sekarang semua umur sudah realistis.
```

```python
# Cek nilai unik di kolom Jenis Kelamin
df['Jenis Kelamin'].unique()

# Insight:
# - Ada nilai 'Alien' → tidak valid
# - Ada penulisan dobel (Laki-laki vs laki-laki, Perempuan vs perempuan)
# - Artinya kolom ini perlu dibersihkan supaya seragam.
```

```python
# Lihat jumlah masing-masing kategori
df['Jenis Kelamin'].value_counts()

# Insight:
# - 'Alien' ada 2 orang
# - 'laki-laki' cuma beda kapitalisasi
# - 'perempuan' juga beda kapitalisasi
# - Kita perlu standarisasi penulisan + hapus kategori tidak valid.
```

```python
# Filter baris dengan 'Alien'
filter_jk = df['Jenis Kelamin'] == 'Alien'
df[filter_jk]

# Insight:
# - Alien ternyata Maradona & Figo
# - Kita asumsikan ini salah input, lalu diganti dengan 'Laki-laki'.
```

```python
# Ganti nilai 'Alien' jadi 'Laki-laki'
index_jk = df[filter_jk].index
df.loc[index_jk, 'Jenis Kelamin'] = 'Laki-laki'
df['Jenis Kelamin'].value_counts()

# Insight:
# - Alien sudah hilang
# - 'Laki-laki' sekarang 6 orang
```

```python
# Perbaiki penulisan supaya konsisten
df['Jenis Kelamin'] = df['Jenis Kelamin'].str.replace('laki-laki', 'Laki-laki')
df['Jenis Kelamin'] = df['Jenis Kelamin'].str.replace('perempuan', 'Perempuan')
df['Jenis Kelamin'].value_counts()

# Insight:
# - Sekarang hanya ada 2 kategori valid:
#   -> 'Laki-laki' (7 orang)
#   -> 'Perempuan' (3 orang)
# - Semua error kategori sudah dibereskan.
```

```python
# Lihat isi DataFrame awal → kolom 'Email dan POS' masih nyampur (email + kode pos digabung dengan koma)
df
```

```python
# Ambil khusus kolom 'Email dan POS' → hasilnya masih berupa 1 kolom object (string)
df['Email dan POS']
```

```python
# Pisahkan string berdasarkan tanda koma (,)
# expand=True → hasil split langsung dibagi ke 2 kolom (kolom sementara tanpa nama)
df['Email dan POS'].str.split(',', expand=True)

# Ambil hasil split kolom pertama (index [0]) → simpan sebagai kolom baru 'Email'
df['Email'] = df['Email dan POS'].str.split(',', expand=True)[0]

# Ambil hasil split kolom kedua (index [1]) → simpan sebagai kolom baru 'POS'
df['POS'] = df['Email dan POS'].str.split(',', expand=True)[1]

# Sekarang sudah ada kolom baru 'Email' & 'POS', tapi kolom lama 'Email dan POS' masih ada
df
```

```python
# Hapus kolom 'Email dan POS' karena sudah tidak dipakai
# axis=1 → berarti menghapus kolom, inplace=True → langsung ubah df tanpa bikin salinan
df.drop('Email dan POS', axis=1, inplace=True)
df
```
