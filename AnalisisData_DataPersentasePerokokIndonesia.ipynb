# Analisis Data Perokok di Indonesia 2021–2024  

Analisis ini menggunakan data dari Badan Pusat Statistik (BPS). Data yang dipakai adalah Persentase Penduduk Berumur 15 Tahun ke Atas yang Merokok Tembakau selama Sebulan Terakhir Menurut Kelompok Umur (Persen) periode 2021–2024. Biar lebih jelas, data mentah tersebut diolah ulang dengan Pandas, pustaka Python untuk analisis data.

```python
pd.options.display.max_rows = 10
df
# Tampilkan data mentah persentase perokok di Indonesia 2021–2024
# Data terdiri dari 44 baris (11 rentang umur x 4 tahun)dan 3 kolom: Umur, Tahun, Persentase Perokok
```

##1. Persentase perokok tertinggi di tahun 2021–2024

```python
# Persentase perokok tertinggi di tahun 2021–2024

# Ambil data persentase tertinggi tiap tahun
persentase_tertinggi = df.loc[df.groupby('Tahun')['Persentase Perokok'].idxmax()]
persentase_tertinggi = persentase_tertinggi.reset_index()                           # reset index
persentase_tertinggi.index += 1                                                     # ubah index mulai dari 1
persentase_tertinggi[['Tahun', 'Umur', 'Persentase Perokok']]

# Insight: kelompok umur 35-39 dan 40-44 cenderung memiliki persentase perokok tertinggi
```

##2. Persentase perokok terendah di tahun 2021–2024

```python
# Persentase perokok terendah di tahun 2021–2024

# Ambil data persentase terendah tiap tahun
persentase_terendah = df.loc[df.groupby('Tahun')['Persentase Perokok'].idxmin()]
persentase_terendah = persentase_terendah.reset_index()                            # reset index
persentase_terendah.index += 1                                                     # ubah index mulai dari 1
persentase_terendah[['Tahun', 'Umur', 'Persentase Perokok']]

# Insight: kelompok umur 15-19 selalu memiliki persentase terendah
```

##3. Tabel persentase perokok per tahun 2021-2024

```python
#Tabel - persentase perokok per tahun (2021-2024)

# Pivot table: baris = umur, kolom = tahun, nilai = persentase perokok
pivot_df = df.pivot(index='Umur', columns='Tahun', values='Persentase Perokok')

# Tambahkan kolom rata-rata persentase perokok tiap umur
pivot_df['Rata-rata'] = pivot_df.mean(axis=1).round(2)

# Tampilkan semua baris dan kolom
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pivot_df

# Pivot table: baris = umur, kolom = tahun, nilai = persentase perokok
# Tujuan: memudahkan melihat tren perokok tiap umur dari 2021–2024
```

##4. Perubahan tahunan setiap umur (dalam persen)

```python
# Perubahan tahunan setiap umur (dalam persen)
perubahan_tahunan = pivot_df[[2021, 2022, 2023, 2024]].diff(axis=1).round(2)
perubahan_tahunan = perubahan_tahunan.fillna(0)
# Diff per tahun: menghitung perubahan persentase perokok tiap umur

# mean perubahan tahunan setiap umur
perubahan_tahunan['mean'] = perubahan_tahunan.mean(axis=1).round(2)
# mean: rata-rata perubahan tahunan tiap umur (positif = naik, negatif = turun)

#standar deviasi untuk mengetahui seberapa besar fluktuasi data
perubahan_tahunan['std'] = perubahan_tahunan[[2021, 2022, 2023, 2024]].std(axis=1).round(2)
# std: standar deviasi perubahan tahunan, menunjukkan fluktuasi data

perubahan_tahunan
# tampilkan data Perubahan tahunan setiap umur
pd.set_option('display.max_rows', None)
perubahan_tahunan
```

```python
# rata-rata perubahan tahunan

# Ambil baris dengan rata-rata perubahan tahunan
mean_tertinggi = perubahan_tahunan.loc[[perubahan_tahunan['mean'].idxmax()]]
mean_terendah = perubahan_tahunan.loc[[perubahan_tahunan['mean'].idxmin()]]
```

```python
# tampilkan rata-rata perubahan tahunan tertinggi
mean_tertinggi
# Umur dengan rata-rata kenaikan persentase perokok tertinggi: 45-49
```

```python
# tampilkan rata-rata perubahan tahunan terendah
mean_terendah
# Umur dengan rata-rata penurunan persentase perokok tertinggi: 30-34
```
