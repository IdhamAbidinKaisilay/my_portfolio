## 2. Buatlah **pie chart** penjualan (`harga per unit` x `bulan_target`) berdasarkan bulan, gunakan variabel `bulan_target` dalam membuat pie chart, jika `bulan_target` diubah, maka **pie chart** juga akan berubah. Buatlah nilai penjualan terbesar menjadi `explode` atau keluar.

```python
bulan_target = 'Januari'

# Ketikan kode anda dibawah sini
label = df['Jenis Bahan Bangunan']
penjualan = df['Harga per Unit (Rp)'] * df[bulan_target]

param_explode = []
max_value = penjualan.max()

for value in penjualan:
  if value == max_value:
    param_explode.append(0.1)
  else:
    param_explode.append(0)

plt.figure(figsize=(6, 6))
plt.pie(penjualan, labels=label,
        explode=param_explode,
        autopct="%.2f%%", pctdistance=0.8,
        textprops={'fontsize':8})

plt.title(f"Penjualan Bulan {bulan_target}")

plt.show()
```

```python
bulan_target = 'Maret'

# Ketikan kode anda dibawah sini
label = df['Jenis Bahan Bangunan']
penjualan = df['Harga per Unit (Rp)'] * df[bulan_target]

param_explode = []
max_value = penjualan.max()

for value in penjualan:
  if value == max_value:
    param_explode.append(0.1)
  else:
    param_explode.append(0)

plt.figure(figsize=(6, 6), facecolor='#888888')
plt.pie(penjualan, labels=label,
        explode=param_explode,
        autopct="%.2f%%", pctdistance=0.8,
        textprops={'fontsize':8,
                   'color':'#FFFFFF',
                   'weight':'light',
                   'style':'oblique'})

plt.title(f"Penjualan Bulan {bulan_target}", color='white')

plt.show()
```

```python
# buat fungsi pie_chart
def pie_chart(bulan_target):
  # Ketikan kode anda dibawah sini
  label = df['Jenis Bahan Bangunan']
  penjualan = df['Harga per Unit (Rp)'] * df[bulan_target]

  param_explode = []
  max_value = penjualan.max()

  for value in penjualan:
    if value == max_value:
      param_explode.append(0.1)
    else:
      param_explode.append(0)

  plt.figure(figsize=(6, 6))
  plt.pie(penjualan, labels=label,
          explode=param_explode,
          autopct="%.2f%%", pctdistance=0.8,
          textprops={'fontsize':8})

  plt.title(f"Penjualan Bulan {bulan_target}")

  plt.show()
```

```python
pie_chart('Oktober')
```
